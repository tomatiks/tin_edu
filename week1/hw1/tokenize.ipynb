```python
import pandas as pd
import maru
import razdel
from tqdm import tqdm
import json
```

```python
maru_rnn = maru.get_analyzer('rnn')

def lemmatize_with_maru(tokens):
    return [morph.lemma for morph in maru_rnn.analyze(tokens)]

def tokenize_with_razdel(text):
    return [token.text for token in razdel.tokenize(text)]
```

```python
train_data = pd.read_csv('./data/train.csv', index_col='id')
train_text = train_data.values[:,0]
train_tokens = [lemmatize_with_maru(tokenize_with_razdel(text)) for text in tqdm(train_text)]
json.dump(train_tokens, open('./data/train_tokens.json', 'w'))
```

```python
test_data = pd.read_csv('./data/test.csv', index_col='id')
test_text = test_data.values[:,0]
test_tokens = [lemmatize_with_maru(tokenize_with_razdel(text)) for text in tqdm(test_text)]
json.dump(test_tokens, open('./data/test_tokens.json', 'w'))
```
